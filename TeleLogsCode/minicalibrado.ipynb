```python
import re
import numpy as np
import pandas as pd
```

```python
# Cargar el archivo
file_path = "calibrados\calibrado-18.txt"
with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

# Expresión regular para extraer los datos
pattern = re.compile(
    r"Iteración (\d+).*?Dt = ([\d.]+) segundos.*?"
    r"Altitud relativa: ([\d\-.]+) m.*?"
    r"Velocidad terrestre: ([\d.]+) m/s.*?"
    r"Pitch \(canal 2\): (\d+), Throttle \(canal 3\): (\d+)",
    re.DOTALL
)

# Buscar todas las coincidencias
matches = pattern.findall(text)

# Crear DataFrame
df = pd.DataFrame(matches, columns=["Iteración", "Dt (s)", "Altitud (m)", "Velocidad (m/s)", "Pitch", "Throttle"])

# Convertir tipos
df = df.astype({
    "Iteración": int,
    "Dt (s)": float,
    "Altitud (m)": float,
    "Velocidad (m/s)": float,
    "Pitch": int,
    "Throttle": int
})
```
